```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Bar Chart with Labels (1.30.1)

```python
job_country = df.job_country.value_counts().head()
plt.bar(job_country.index, job_country, color='green')
plt.xlabel('Country')
plt.ylabel('Number of Postings')
plt.title('Number of Jobs per Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

---

## ðŸŸ¨ Horizontal Bar Chart with Labels (1.30.2)

```python
average_yearly_job_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
average_yearly_job_salary = average_yearly_job_salary.sort_values()
plt.barh(average_yearly_job_salary.index, average_yearly_job_salary, color='red')
plt.xlabel('Average Salary')
plt.title('Average Salary by Job Title')
plt.show()
```

---

## ðŸŸ¥ Line Chart with Labels (1.30.3)

```python
df['job_posted_month'] = df.job_posted_date.dt.month
avg_salary_per_month = df.groupby('job_posted_month')['salary_year_avg'].mean()
plt.plot(avg_salary_per_month.index, avg_salary_per_month, color='red', label='Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by job Month')
plt.legend()
plt.show()
```
