```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df
```

## 🟩 Concat Job Postings (2.8.1)

```python
data_scientist_df = df[df['job_title_short'] == 'Data Scientist']
data_analyst_df = df[df['job_title_short'] == 'Data Analyst']

df_concat = pd.concat([data_scientist_df, data_analyst_df], ignore_index=True)
df_concat
```

---

## 🟨 Concat Monthly Postings (2.8.2)

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

months = ['paź', 'lis', 'gru']

quarter_df_dict = {f'{month}_df': df[df['job_posted_month'] == month] for month in months}
quarter_df_dict['paź_df']

combined_df = pd.concat([quarter_df_dict['paź_df'], quarter_df_dict['lis_df'], quarter_df_dict['gru_df']], ignore_index=True)
combined_df

#Luke
# oct_df = df[df['job_posted_date'].dt.month == 10]
# nov_df = df[df['job_posted_date'].dt.month == 11]
# dec_df = df[df['job_posted_date'].dt.month == 12]
# combined_df = pd.concat([oct_df, nov_df, dec_df], ignore_index=True)
# combined_df

```

## 🟨 Concat Job Types (2.8.3)

```python
full_time_df = df[df['job_schedule_type'] == 'Full-time']
part_time_df = df[df['job_schedule_type'] == 'Part-time']

combined_df = pd.concat([full_time_df, part_time_df], ignore_index=True)
combined_df
```
