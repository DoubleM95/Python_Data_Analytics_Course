```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Remove Rows with Missing Values (2.2.1)

```python
drop_df = df.dropna(subset='salary_year_avg')
print('Original df: ', len(df))
print('Dropped df: ', len(drop_df))
print('Difference: ', len(df) - len(drop_df))
```

---

## ðŸŸ¨ Remove Duplicate Rows (2.2.2)

```python
drop_df  = df.drop_duplicates(subset='job_location')
print('Original df: ', len(df))
print('Dropped df: ', len(drop_df))
print('Difference: ', len(df) - len(drop_df))
```
