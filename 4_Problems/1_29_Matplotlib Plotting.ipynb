```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Country Bar Chart (1.29.1)

```python
country_jobs = df.job_country.value_counts().head()
plt.bar(country_jobs.index, country_jobs)
plt.show()
```

---

## ðŸŸ¨ Average Salary Horizontal Bar Chart (1.29.2)

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
avg_salary = avg_salary.sort_values(ascending=True)
plt.barh(avg_salary.index, avg_salary)
plt.show()
```

---

## ðŸŸ¥ Average Salary Trend Line Plot (1.29.3)

```python
df['job_posted_month'] = df.job_posted_date.dt.month

monthly_avg_salary = df.groupby('job_posted_month')['salary_year_avg'].mean()
plt.plot(monthly_avg_salary.index, monthly_avg_salary)

```
